```python
class LinkedListDict:
    def __init__(self):
        self.head = None  # Head stores the key of the first node
        self.nodes = {}   # Dictionary to store nodes

    # Method to append a node at the end
    def append(self, key, data):
        new_node = {'data': data, 'next': None}
        self.nodes[key] = new_node
        if self.head is None:
            self.head = key
        else:
            # Traverse to the last node
            current_key = self.head
            while self.nodes[current_key]['next'] is not None:
                current_key = self.nodes[current_key]['next']
            self.nodes[current_key]['next'] = key

    # Method to display the linked list
    def display(self):
        current_key = self.head
        while current_key is not None:
            print(self.nodes[current_key]['data'], end=" -> ")
            current_key = self.nodes[current_key]['next']
        print("None")

    # Method to delete a node by key
    def delete(self, key):
        if key not in self.nodes:
            print(f"{key} not found in the list")
            return
        if key == self.head:
            self.head = self.nodes[key]['next']
        else:
            # Find the previous node
            current_key = self.head
            while current_key and self.nodes[current_key]['next'] != key:
                current_key = self.nodes[current_key]['next']
            if current_key:
                self.nodes[current_key]['next'] = self.nodes[key]['next']
        # Remove node from dictionary
        del self.nodes[key]

# Example usage
ll = LinkedListDict()
ll.append('a', 10)
ll.append('b', 20)
ll.append('c', 30)

ll.display()  # Output: 10 -> 20 -> 30 -> None

ll.delete('b')
ll.display()  # Output: 10 -> 30 -> None
```
